```python
#aggregated transaction
import pandas as pd
import json
import os

path=r'G:/phonepe/data/data/aggregated/transaction/country/india/state/'
Agg_state_list=os.listdir(path)
Agg_state_list
print(Agg_state_list)

clm={'State':[], 'Year':[],'Quarter':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            with open(p_k,'r') as Data:
                D=json.load(Data)
                for z in D['data']['transactionData']:
                    clm['Transaction_type'].append(z['name'])
                    clm['Transaction_count'].append(z['paymentInstruments'][0]['count'])
                    clm['Transaction_amount'].append(z['paymentInstruments'][0]['amount'])
                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quarter'].append(int(k.strip('.json')))

aggregated_transaction = pd.DataFrame(clm)

# Show result
print(aggregated_transaction.head())
print("Rows:", len(aggregated_transaction))
```

```python
#aggregated user

import pandas as pd
import json
import os

# Correct folder path
path = r"/phonepe/data/data/aggregated/user/country/india/state/"

# Get list of states
state_list = os.listdir(path)
print(state_list)

# Create empty dictionary
clm = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Brand': [],
    'Count': [],
    'Percentage': []
}

# Loop through states
for state in state_list:
    state_path = os.path.join(path, state)
    
    if os.path.isdir(state_path):
        years = os.listdir(state_path)
        
        # Loop through years
        for year in years:
            year_path = os.path.join(state_path, year)
            
            if os.path.isdir(year_path):
                quarters = os.listdir(year_path)
                
                # Loop through quarter files
                for quarter_file in quarters:
                    if quarter_file.endswith('.json'):
                        file_path = os.path.join(year_path, quarter_file)
                        
                        with open(file_path, 'r') as f:
                            data = json.load(f)
                            
                            try:
                                users = data["data"]["usersByDevice"]
                                
                                for user in users:
                                    clm['State'].append(state)
                                    clm['Year'].append(int(year))
                                    clm['Quarter'].append(int(quarter_file.strip('.json')))
                                    clm['Brand'].append(user['brand'])
                                    clm['Count'].append(user['count'])
                                    clm['Percentage'].append(user['percentage'])
                            
                            except (KeyError, TypeError):
                                # Some files may not contain usersByDevice
                                pass

# Create DataFrame
aggregated_user = pd.DataFrame(clm)

# Show result
print(aggregated_user.head())
print("Rows:", len(aggregated_user))
```
